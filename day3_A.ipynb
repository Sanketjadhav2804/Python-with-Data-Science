### Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.shape
```

```python
df
```

```python
plt.scatter(df['YearsExperience'] , df['Salary'])
```

#### seprate the input and output datam

```python
x= df['YearsExperience'].values 
y =df['Salary']
```

```python
x
```

```python
y
```

```python
x = x.reshape(30,1)
```

```python
x
```

### import linear regression class


```python
from sklearn.linear_model import LinearRegression
```

### create objject of linear regression 

```python
regressor = LinearRegression()
```

### train the algorithm

```python
regressor.fit(x, y)
```

### predict on unseen data

```python
regressor.predict([[15]])
```

```python
regressor.predict([[0]])
```

```python
regressor.score(x,y)
```
